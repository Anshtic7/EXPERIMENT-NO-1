AREA OF CIRCLE
where r is the radius taken by the user


```python
import math
radius = float(input("Enter the radius of the circle: "))
area = math.pi * (radius ** 2)
print(f"The area of the circle with radius {radius} is: {area:.2f}")
```

VALUE OF LOG 10 BASE OF 10


```python
import math
result = math.log10(10)
print(f"The value of log base 10 of 10 is: {result}")
```

VALUE OF SIN 60

```python
import math
degrees = 60
radians = math.radians(degrees)
sin_value = math.sin(radians)
print(f"The sine of {degrees} degrees is: {sin_value}")
```

USING IF ELSE TO FIND GREATER NUMBER (also taking input from the user)

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
if a > b:
  print("a is greater than b")
else:
  print("b is greater than a")
```
